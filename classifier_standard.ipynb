```python
import turicreate as tc
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
def create_sframe(path, name):
    data = tc.image_analysis.load_images(path, with_path=True)
    data["label"] = data["path"].apply(lambda path: "busy" if "/busy" in path else "not_busy")
    data.save(name + ".sframe")
    data.show()
```

```python
create_sframe("datasets/train6/", "train6")
create_sframe("datasets/train5/", "train5")
create_sframe("datasets/train4/", "train4")
create_sframe("datasets/train3/", "train3")
create_sframe("datasets/train2/", "train2")
create_sframe("datasets/test/", "test")

train6_data = tc.SFrame("train6.sframe")
train5_data = tc.SFrame("train5.sframe")
train2_data = tc.SFrame("train2.sframe")
test_data = tc.SFrame("test.sframe")
```

```python
model6 = tc.image_classifier.create(train6_data, target="label", model="resnet-50")
model5 = tc.image_classifier.create(train5_data, target="label", model="resnet-50")
model2 = tc.image_classifier.create(train5_data, target="label", model="resnet-50")

predictions6 = model6.predict(test_data)
predictions5 = model5.predict(test_data)
predictions2 = model2.predict(test_data)
```

```python
metrics6 = model6.evaluate(test_data)
metrics5 = model5.evaluate(test_data)
metrics2 = model2.evaluate(test_data)
```

```python
def show_metrics(metrics):
    accuracy = metrics["accuracy"]
    precision = metrics["precision"]
    recall = metrics["recall"]
    f1 = metrics["f1_score"]
    confusion_matrix = metrics["confusion_matrix"]
    roc_curve = metrics["roc_curve"]

    print(f"accuracy: {accuracy}")
    print(f"precision: {precision}")
    print(f"recall: {recall}")
    print(f"f1: {f1}")
    print(confusion_matrix)

    plt.plot([0, 1], [0, 1], color='orange', linestyle='--')
    plt.plot(roc_curve["fpr"], roc_curve["tpr"])
```

```python
show_metrics(metrics6)
show_metrics(metrics5)
show_metrics(metrics2)
```
